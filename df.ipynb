```python
%%writefile mobilenet_model.py
import torch
import torch.nn as nn
from torchvision.models import mobilenet_v2

class DeepFakeMobileNet(nn.Module):
    def __init__(self):
        super(DeepFakeMobileNet, self).__init__()
        self.model = mobilenet_v2(weights="IMAGENET1K_V1")
        self.model.classifier[1] = nn.Linear(self.model.last_channel, 2)

    def forward(self, x):
        return self.model(x)

def load_model():
    model = DeepFakeMobileNet()
    model.eval()
    return model
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import torch
from mobilenet_model import load_model

model = load_model()
model.eval()

MEAN = np.array([0.485, 0.456, 0.406], dtype=np.float32)
STD  = np.array([0.229, 0.224, 0.225], dtype=np.float32)

def preprocess(frame_bgr):
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(rgb, (224, 224))
    img = resized.astype(np.float32) / 255.0
    img = (img - MEAN) / STD
    img = np.transpose(img, (2, 0, 1))
    return torch.from_numpy(img).unsqueeze(0)

def predict_label(frame_bgr):
    x = preprocess(frame_bgr)
    with torch.no_grad():
        out = model(x)
        _, pred = torch.max(out, 1)
    return "FAKE" if pred.item() == 1 else "REAL"

st.title("Deepfake Detection ‚Äî Samartha (BE AIML)")
mode = st.sidebar.selectbox("Mode", ["Image Detection", "Video Verdict"])

if mode == "Image Detection":
    file = st.file_uploader("Upload image", type=["jpg", "jpeg", "png"])
    if file:
        data = np.asarray(bytearray(file.read()), dtype=np.uint8)
        img = cv2.imdecode(data, 1)
        label = predict_label(img)
        st.image(cv2.cvtColor(img, cv2.COLOR_BGR2RGB),
                 caption=f"Prediction: {label}")

elif mode == "Video Verdict":
    file = st.file_uploader("Upload video", type=["mp4", "avi"])
    if file:
        path = "video.mp4"
        with open(path, "wb") as f:
            f.write(file.read())
        cap = cv2.VideoCapture(path)
        real=fake=0
        total=0
        while True:
            ret, frame = cap.read()
            if not ret: break
            total += 1
            if total % 10 != 0: continue
            if predict_label(frame) == "REAL": real += 1
            else: fake += 1
        cap.release()
        final = "REAL" if real > fake else "FAKE"
        st.subheader(f"FINAL VERDICT: {final}")
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import torch
from mobilenet_model import load_model
from PIL import Image

st.set_page_config(page_title="Deepfake Detector", layout="centered")
st.title("Deepfake Detection - Samartha (BE AIML)")
st.write("Upload Image or Video to check if it's REAL or FAKE")

model = load_model()
model.eval()

MEAN = np.array([0.485, 0.456, 0.406], dtype=np.float32)
STD = np.array([0.229, 0.224, 0.225], dtype=np.float32)

def preprocess(bgr):
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(rgb, (224, 224))
    img = resized.astype(np.float32)/255.0
    img = (img - MEAN) / STD
    img = np.transpose(img, (2,0,1))
    return torch.from_numpy(img).unsqueeze(0)

def classify(frame):
    x = preprocess(frame)
    with torch.no_grad():
        out = model(x)
        _, pred = torch.max(out, 1)
    return "FAKE" if pred.item() == 1 else "REAL"

mode = st.sidebar.radio("Select Option", ["Image", "Video"])

if mode == "Image":
    file = st.file_uploader("Upload Image", type=["jpg", "jpeg", "png"])
    if file:
        bytes = np.asarray(bytearray(file.read()), dtype=np.uint8)
        frame = cv2.imdecode(bytes, 1)
        pred = classify(frame)
        st.image(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB),
                 caption=f"Prediction: {pred}",
                 use_column_width=True)

elif mode == "Video":
    file = st.file_uploader("Upload Video", type=["mp4", "avi"])
    if file:
        video_path = "uploaded_video.mp4"
        with open(video_path, "wb") as f:
            f.write(file.read())

        st.info("Analyzing video... Please wait ‚è≥")
        cap = cv2.VideoCapture(video_path)

        total = real = fake = 0
        while True:
            ret, frame = cap.read()
            if not ret: break
            total += 1
            if total % 10 != 0: continue

            if classify(frame) == "REAL": real += 1
            else: fake += 1

        cap.release()

        final = "REAL" if real > fake else "FAKE"
        st.success(f"üéØ Final Verdict: **{final}**")
        st.caption("Note: Prototype Model - To be trained on Deepfake Dataset")
```
